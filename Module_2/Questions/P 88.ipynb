# Q.1
I am recreating the PDF and CDF for student's t distribution

```python
from numpy import arange
from matplotlib import pyplot
from scipy.stats import t
SS = arange(-9,10,0.001)
dof = 3
pdf = t.pdf(SS,dof)
pyplot.plot(SS,pdf)
pyplot.show()
```

```python
cdf = t.cdf(SS,dof)
pyplot.plot(SS,cdf)
pyplot.show()
```

# Q.2

## Cauchy Distribution.

```python
from scipy.stats import cauchy
SS = arange(-4,4,0.001)
pdf = cauchy.pdf(SS)
pyplot.plot(SS,pdf)
pyplot.show()
cdf = cauchy.cdf(SS)
pyplot.plot(SS,cdf)
pyplot.show()
```

## Laplace Distribution.

```python
from scipy.stats import laplace
SS = arange(-10,10,0.001)
pdf = laplace.pdf(SS)
pyplot.plot(SS,pdf)
pyplot.show()
cdf = laplace.cdf(SS)
pyplot.plot(SS,cdf)
pyplot.show()
```

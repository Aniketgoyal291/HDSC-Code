# Plotting gaussian function

```python
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm
x_axis = arange(-3,3,0.001)
y_axis = norm.pdf(x_axis, 0,1)
pyplot.plot(x_axis, y_axis)
pyplot.show()
```

# Generating random numbers from gaussian distribution.

```python
from numpy.random import randn, seed
from matplotlib import pyplot
# Seeding the random number generator
seed(1)
data = 5*randn(10000)+50
pyplot.hist(data)
pyplot.show()
```

# Creating splits in the above graph

```python
pyplot.hist(data, bins = 100)
pyplot.show()
```

# Calculating Mean and Median and Variance.

```python
import numpy as np
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
print('Mean of the data is:', mean)
print('Median of the data is:', median)
print('Variance of the data is:', variance)
```

# Gaussian Dist. with different variance.

```python
x_axis = arange(-3,3,0.001)
# Low variance   (BLUE)
pyplot.plot(x_axis, norm.pdf(x_axis, 0,0.5))
# High variance
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 1))
pyplot.show()
```

# Line Plot

# Bar Chart

```python
from random import seed, randint
from matplotlib import pyplot
seed(2)
X = ['red', 'green', 'blue']
Y = [randint(0,100),randint(0,100),randint(0,100)]
pyplot.bar(X,Y)
pyplot.show()
```

# Histogram Plot

```python
from numpy.random import seed, randn
from matplotlib import pyplot as pl
seed(1)
data = randn(1000)
pl.hist(data)
pl.show()
```

# Scatter plot

```python
from numpy.random import seed, randn
from matplotlib import pyplot
seed(3)
X = 20*randn(1000) + 100
Y = X + (10*randn(1000) + 50)
pyplot.scatter(X,Y)
pyplot.show()
```

# Demonstration of law of large numbers


```python
# Idealized distribution With Mean = 50 And Stdev = 5
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm
X = arange(30,70,1)
Y = norm.pdf(X,50,5)
pyplot.plot(X,Y)
pyplot.show()
```

```python
# Now we will plot difference of mean vs the size of data taken.

from numpy.random import seed, randn
from matplotlib import pyplot
from numpy import array, mean
seed(1)
Sizes = list()
for x in range(10,20000,200):
    Sizes.append(x)
Means = [mean(5*randn(size)+50) for size in Sizes]
pyplot.scatter(Sizes, array(Means) - 50)
pyplot.show()
# We converted Means into np array because we cannot perform subtraction in list.
```

# Example of Central Limit Theorm
We will take an example of rolling of dice and understand CLT with it.

```python
from numpy.random import seed, randint
from numpy import mean
seed(1)
rolls = randint(1,7,50)
print(rolls)
print(mean(rolls))
```

```python
means = [mean(randint(1,7,1000)) for i in range(1000)]
pyplot.hist(means)
pyplot.show()
```

# Distributions

## (1) Gaussian Distribution

```python
from numpy import arange
from matplotlib import pyplot 
from scipy.stats import norm
sample_space = arange(-5,5,0.001)
pdf = norm.pdf(sample_space,0,1)
pyplot.plot(sample_space,pdf)
pyplot.show()
cdf = norm.cdf(sample_space)
pyplot.plot(sample_space,cdf)
pyplot.show()
```

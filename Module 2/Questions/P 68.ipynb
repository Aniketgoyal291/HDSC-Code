```python
from numpy.random import seed, randint
from numpy import mean
from matplotlib import pyplot
seed(1)
for i in range(100,100000,200):
    A = [mean(randint(1,7,i))]
    for x in range(len(A)):
        B = mean(A[x])
        pyplot.scatter(i,B)
```

As we can see in this above code, as we increase the size of sample taken from population, the graph closes in on our expected mean value i.e. 3.5

```python
from matplotlib import pyplot
from numpy import mean
from numpy.random import randn
def gaussian_numbers():
    means = [mean(randn(1000)) for a in range(1000)]
    pyplot.hist(means)
    pyplot.show()
    
    
gaussian_numbers()
```

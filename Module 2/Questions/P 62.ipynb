```python
import numpy as np
from matplotlib import pyplot
from scipy.stats import norm
X = np.arange(10,90)
Y = norm.pdf(X,45,10)
pyplot.plot(X,Y)
pyplot.show()
```

```python
from numpy.random import randn
from matplotlib import pyplot
sizes = []
for i in range(100,10000,200):
    sizes.append(i)
mean = [np.mean(10*randn(size)+45) for size in sizes]
pyplot.scatter(sizes, np.array(mean) - 45)
pyplot.show()
```

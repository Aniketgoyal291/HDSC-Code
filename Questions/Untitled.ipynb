```python
import numpy as np
from matplotlib import pyplot
from numpy.linalg import inv
data = np.array([[1,2],[2,4],[3,8],[7,24]])
X, Y = data[:,0],data[:,1]
X = X.reshape((len(X),1))
pyplot.scatter(X,Y)
pyplot.show()
```

```python
b = inv(X.T.dot(X)).dot(X.T).dot(Y)
print(b)
```

```python
yhat = X.dot(b)
```

```python
pyplot.scatter(X,Y)
pyplot.plot(X,yhat,color = 'green')
pyplot.show()
```
